# STT707

Soit une matrice de $n$ observations comportant $p$ mesures chacune
$
X_{n \times p} \equiv \{ x_i^j \}, \,\,\, i = 1,2,...,n, \,\,\, j=1,2,...,p
$

$
X = \begin{Bmatrix}
    x_{1}^{1} & x_{1}^{2} & \cdots & x_{1}^{p} \\
    x_{2}^{1} & x_{2}^{2} & \cdots & x_{2}^{p} \\
    \vdots & \vdots &\ddots & \vdots \\
    x_{n}^{1} &x_{n}^{2} & \cdots & x_{n}^{p} \\
    \end{Bmatrix}
$

```python
# fixons les dimensions
n = 10 # Observations
p = 3 # so we can represent the data in a 3D space

X = np.random.rand(n, p) * 10
# print(f'{X[:5, :]}')
# print('\t.\t.\t.\n'*3)
# print(f'{X[-5:, :]}')
```

```python
fig = plt.figure()
ax = fig.gca(projection='3d')
ax.scatter(X[:, 0], X[:, 1], X[:, 2])
```

```python
fig1 = plt.figure(figsize=(15, 15))
ax = fig1.gca(projection='3d')

ax.scatter(X[:, 0], X[:, 1], X[:, 2], c='b')
for n in range(0, X.shape[0]):
    _vec = np.array([[*np.zeros(X.shape[1])], [*X[n, :]]])
    ax.plot(_vec[:, 0], _vec[:, 1], _vec[:, 2], c='b')
ax.text(x=-0.1, y=-0.1, z=-0.1, s='O')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
```

Soit $A$ un matrice diagonale dfinie positive. Nous avons les relations suivantes:

$
A = \begin{bmatrix}
        a_1 & 0     & \dots & 0 \\
        0   & a_2   & \dots & 0  \\
        \vdots & \vdots & \ddots & \vdots \\
        0 & 0 & \dots & a_p
    \end{bmatrix} \\
$

$
\begin{equation}
    \begin{split}
            a_i &= Ax_i \\
            &= \begin{bmatrix}
                a_1 & 0     & \dots & 0 \\
                0   & a_2   & \dots & 0  \\
                \vdots & \vdots & \ddots & \vdots \\
                0 & 0 & \dots & a_p
            \end{bmatrix}
            \begin{bmatrix}
                x_i^1   \\
                x_i^2   \\
                \vdots   \\
                x_i^p   \\
            \end{bmatrix}
            &= \sum_{j=1}^p a_j x_i^j
    \end{split}
\end{equation}
$


```python
# A = np.identity(X.shape[0]) * -.5
A = np.diag(np.linspace(1, X.shape[0], X.shape[0]))
fig = plt.figure(figsize=(15, 15))
ax = fig.gca(projection='3d')
# ax = fig.gca()

X_star = X.copy()
X_star = np.dot(A, X_star)
ax.scatter(X_star[:, 0], X_star[:, 1], X_star[:, 2], c='r')
ax.scatter(X[:, 0], X[:, 1], X[:, 2], c='b')
for n in range(0, X_star.shape[0]):
    _vec = np.array([[*np.zeros(X_star.shape[1])], [*X_star[n, :]]])
    ax.plot(_vec[:, 0], _vec[:, 1], _vec[:, 2], c='r')
    _vec = np.array([[*np.zeros(X.shape[1])], [*X[n, :]]])
    ax.plot(_vec[:, 0], _vec[:, 1], _vec[:, 2], c='b')

ax.text(x=-0.1, y=-0.1, z=-0.1, s='O')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
```
